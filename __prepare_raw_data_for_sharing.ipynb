```python
import numpy as np
import numpy.ma as ma
import h5py
import scipy.io
import matplotlib.pyplot as plt
import os
```

```python
fg = f['data']
marker = np.array(f['marker_labels']) 
labels = list(f['data'].keys())

allowed_prefixes = {'HC', 'OFF', 'ON', 'Single ring', 'OFF DBS'}

rename_map = {
    'ALb': 'PD01',
    'CAe': 'PD02',
    'DEp': 'PD03',
    'DEv': 'PD04',
    'FEp': 'PD05',
    'FRa': 'PD06',
    'FRj': 'PD07',
    'GAl': 'PD08',
    'GUg': 'PD09',
    'HOt': 'PD10',
    'LEe': 'PD11',
    'MAa': 'PD12',
    'PAn': 'PD13',
    'SEc': 'PD14',
    'SOh': 'PD15',
    'TAd': 'PD16',
    'VAj': 'PD17',
    'VIj': 'PD18',
    'BARNI44': 'HC01',
    'BASGI37': 'HC02',
    'BENCH34': 'HC03',
    'BIBTH15': 'HC04',
    'CLAPH41': 'HC05',
    'DANPA32': 'HC06',
    'FARER31': 'HC07',
    'FLODO07': 'HC08',
    'GRAPH36': 'HC09',
    'LATPH43': 'HC10'
}

rename_cond_map = {
    'OFF': 'OFF DOPA',
    'ON': 'ON DOPA',
    'OFF DBS': 'OFF DBS',
    'Single ring': 'ON DBS'
}

filtered_list = []
filtered_keys = []

for key in labels:
    # split only once, safe even if second part contains ';'
    first, second = key.split(';', 1)

    if first in allowed_prefixes:
        second = rename_map.get(second, second)
        first = rename_cond_map.get(first, first)

        new_label = f"{first};{second}"

        filtered_list.append(new_label)
        filtered_keys.append(key)
```

```python
with h5py.File("raw_gait_cycle_kinematics.h5", "w") as ff:
    data_grp = ff.create_group("data")

    for key, new_name in zip(filtered_keys, filtered_list):
        fg.copy(key,data_grp,name=new_name)
    ff.create_dataset('marker_labels', data=np.array(marker, dtype='S'))  # store as byte strings
```
