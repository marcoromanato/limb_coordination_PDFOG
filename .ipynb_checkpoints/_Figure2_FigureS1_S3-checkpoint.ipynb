```python
import pandas as pd
import numpy as np
import os

# --- Load data ---
commonpath = "_st_parameters/"
filename = "_df_spatiotemporal_pd.xlsx"
filename_control = "_df_spatiotemporal_controls.xls"

Imputation = True  # equivalent of Imputation = T

df_long = pd.read_excel(commonpath+filename)
df_ctl = pd.read_excel(commonpath+filename_control)

# --- Drop GONOGO column if present ---
for df in [df_long, df_ctl]:
    if "GONOGO" in df.columns:
        df.drop(columns=["GONOGO"], inplace=True)

# --- Convert numeric columns to numeric, replace NaN or complex values ---
numeric_cols = df_long.loc[:, "Tps_DemiTour":"StepWidth_asym"].columns

for df in [df_long, df_ctl]:
    df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
    # Drop complex numbers (not usual in pandas)
    df = df.replace([np.inf, -np.inf], np.nan)

# --- Remove cycles with wrong DA computations due to FOG episodes ---
mask_wrong = (df_long["DApourcent_R"] < 0) | (df_long["DApourcent_L"] < 0)
df_long = df_long.loc[~mask_wrong].copy()
df_ctl = df_ctl.loc[~mask_wrong].copy()

# --- Keep only trials without FOG ---
df_noFog = (
    df_long.groupby(["Patient", "Session", "Cond", "TrialNum"])
    .filter(lambda x: (x["num_FOG"].iloc[0] == 0) and all(x["num_FOG"].isna() | (x["num_FOG"] == 0)))
)

# Force-include rows for the PAn/OFF_Stim/M6 as has FOG episodes only during turning
subjects_keep = ['PD18', 'PD13']
conditions_keep = ['OFF_Stim', 'OFF_Stim']
sessions_keep = ['M6', 'M6']
extra_rows = df_long[
    (df_long["Patient"].isin(subjects_keep)) &
    (df_long["Cond"].isin(conditions_keep)) &
    (df_long["Session"].isin(sessions_keep))
]
df_noFog = pd.concat([df_noFog, extra_rows]).drop_duplicates().reset_index(drop=True)

# --- Aggregation helper function ---
def summarise_trials(df):
    out = (
        df.groupby(["Patient", "Session", "Cond"], as_index=False)
        .agg(
            Swing_mean = ("Swing_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Swing_L"]]))),
            Swing_sd   = ("Swing_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "Swing_L"]] ))),
            Swing_asym = ("Swing_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "Swing_R"] / df.loc[x.index, "Swing_L"])))),
            
            Stance_mean = ("Stance_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Stance_L"]]))),
            Stance_sd   = ("Stance_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "Stance_L"]] ))),
            Stance_asym = ("Stance_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "Stance_R"] / df.loc[x.index, "Stance_L"])))),
            
            StepTime_mean = ("StepTime_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "StepTime_L"]]))),
            StepTime_sd   = ("StepTime_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "StepTime_L"]] ))),
            StepTime_asym = ("StepTime_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "StepTime_R"] / df.loc[x.index, "StepTime_L"])))),
            
            StepLength_mean = ("Length_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Length_L"]]))),
            StepLength_sd   = ("Length_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "Length_L"]] ))),
            StepLength_asym = ("Length_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "Length_R"] / df.loc[x.index, "Length_L"])))),
            
            Width_mean = ("Width_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Width_L"]]))),
            Width_sd   = ("Width_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "Width_L"]] ))),
            Width_asym = ("Width_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "Width_R"] / df.loc[x.index, "Width_L"])))),
            
            GaitSpeed_mean = ("Vitesse_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Vitesse_L"]]))),
            GaitSpeed_sd   = ("Vitesse_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "Vitesse_L"]] ))),
            GaitSpeed_asym = ("Vitesse_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "Vitesse_R"] / df.loc[x.index, "Vitesse_L"])))),
            
            DoubleS_mean = ("DA_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "DA_L"]]))),
            DoubleS_sd   = ("DA_R", lambda x: np.nanstd(np.concatenate([x, df.loc[x.index, "DA_L"]] ))),
            DoubleS_asym = ("DA_R", lambda x: np.nanmean(np.abs(100 * np.log(df.loc[x.index, "DA_R"] / df.loc[x.index, "DA_L"])))),
            
            Cadence = ("Cadence_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "Cadence_L"]]))),
            num_step_mean = ("num_step_R", lambda x: np.nanmean(np.concatenate([x, df.loc[x.index, "num_step_L"]]))),
            CycleTime = ("Cycle_Time", "mean")
        )
    )
    out = out.rename(columns={"Patient": "Subject", "Cond": "Condition"})
    return out

# --- Summarize for patients and controls ---
df_summary = summarise_trials(df_noFog)
df_summary_ctl = summarise_trials(df_ctl)

# --- Replace infinite values with NaNs for inputation ---
df_summary.replace([np.inf, -np.inf], np.nan, inplace=True)

# --- Imputation for subject PD18 (PREOP, ON DOPA) ---
mask = (df_summary["Subject"] == "PD18") & (df_summary["Session"] == "PREOP") & (df_summary["Condition"] == "ON")
mean_swing_asym = df_summary.loc[df_summary["Subject"] == "PD18", "Swing_asym"].mean(skipna=True)
mean_stance_asym = df_summary.loc[df_summary["Subject"] == "PD18", "Stance_asym"].mean(skipna=True)

df_summary.loc[mask, "Swing_asym"] = mean_swing_asym
df_summary.loc[mask, "Stance_asym"] = mean_stance_asym

# --- Imputation for subject PD16 (M6, ON DBS) ---
mask = (df_summary["Subject"] == "PD16") & (df_summary["Session"] == "M6") & (df_summary["Condition"] == "Single ring")
mean_doubles_asym = df_summary.loc[df_summary["Subject"] == "PD16", "DoubleS_asym"].mean(skipna=True)

df_summary.loc[mask, "DoubleS_asym"] = mean_doubles_asym

# --- Convert all summary numeric columns and filter out non-finite values ---
num_cols = df_summary.columns.difference(["Subject", "Session", "Condition"])

df_summary[num_cols] = df_summary[num_cols].apply(pd.to_numeric, errors="coerce")
df_summary = df_summary[np.isfinite(df_summary[num_cols]).all(axis=1)]

df_summary_ctl[num_cols] = df_summary_ctl[num_cols].apply(pd.to_numeric, errors="coerce")
df_summary_ctl = df_summary_ctl[np.isfinite(df_summary_ctl[num_cols]).all(axis=1)]

# --- Define variable lists ---
qualvar_names = ["Subject", "Session", "Condition"]
quantvar_names = [c for c in df_summary.columns if c not in qualvar_names]
```

```python
def permutation_test(x, y, n_permutations=10000, paired=True, seed=None):
    """Permutation test: paired or independent samples."""
    rng = np.random.default_rng(seed)
    if paired:
        observed = np.mean(x - y)
    else:
        observed = np.mean(x) - np.mean(y)
    
    perm_diffs = np.zeros(n_permutations)
    for i in range(n_permutations):
        if paired:
            signs = rng.choice([1, -1], size=len(x))
            perm_diffs[i] = np.mean((x - y) * signs)
        else:
            pooled = np.hstack([x, y])
            rng.shuffle(pooled)
            
            perm_diffs[i] = np.mean(pooled[:len(x)]) - np.mean(pooled[len(x):])
    
    return np.mean(np.abs(perm_diffs) >= np.abs(observed))

def get_sigstar(p):
    """Return stars depending on p-value."""
    if p < 0.001: return '***'
    elif p < 0.01: return '**'
    elif p < 0.05: return '*'
    else: return 'n.s.'
```

```python
from matplotlib.colors import ListedColormap
import seaborn as sns
from statsmodels.stats.multitest import multipletests

sns.set_theme(context="paper", style="whitegrid", font="Arial")
plt.rcParams.update({
    "axes.linewidth": 1,
    "axes.labelsize": 16,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "legend.fontsize": 12
})

RPCs = ['RPC.1 - Pace', 'RPC.2 - Rhythm', 'RPC.3 - Balance', 'RPC.4', 'RPC.5']

# --- Define your npj-style color palette ---
palette = ['#b91c1c','#fca5a5','#0e7490','#67e8f9']  # light red, dark red, dark teal, light teal

for kd in range(len(RPCs)): 
    # --- Prepare data ---
    data = np.vstack([
        Xp_off_scores[:, kd],     # 0: OFF
        Xp_on_scores[:, kd],      # 1: ON
        Xp_off_dbs_scores[:, kd], # 2: DBS OFF
        Xp_on_dbs_scores[:, kd]   # 3: DBS ON
    ])
    n_conditions, n_subjects = data.shape

    # --- HC mean and SEM ---
    hc_values = data_array_HS[:, kd]
    hc_mean = hc_values.mean()
    hc_sem = np.std(hc_values, ddof=1) / np.sqrt(hc_values.size)

    # --- Create figure ---
    fig, ax = plt.subplots(figsize=(3.5, 3.5))
    x = np.arange(n_conditions)

    # --- Within-subject paired lines ---
    for subj in range(n_subjects):
        ax.plot([0, 1], data[[0, 1], subj], c='gray', alpha=0.25, lw=1, marker='o', markersize=5)
        ax.plot([2, 3], data[[2, 3], subj], c='gray', alpha=0.25, lw=1, marker='o', markersize=5)

    # --- Mean values (colored dots) ---
    mean_vals = data.mean(axis=1)
    for xi, (mean_y, c) in enumerate(zip(mean_vals, palette)):
        ax.scatter(xi, mean_y, marker='o', s=200, color=c, edgecolor='none', zorder=3)

    # --- HC reference (shaded band) ---
    # ax.axhspan(hc_mean - hc_sem, hc_mean + hc_sem, color='black', alpha=0.2, zorder=0)
    ax.axhline(hc_mean, color='black', linestyle='--', alpha=0.3, lw=2)

    # --- X labels ---
    ax.set_xticks([0.5, 2.5])
    ax.set_xticklabels(['PRE-OP', 'POST-OP'], fontsize=14)
    ax.tick_params(axis='y', labelsize=14)
    
    # --- Remove top/right spines for clarity ---
    sns.despine(ax=ax)

    # --- Run permutation tests (as before) ---
    p_off_on  = permutation_test(Xp_off_scores[:, kd], Xp_on_scores[:, kd], paired=True)
    p_off_dbs = permutation_test(Xp_off_dbs_scores[:, kd], Xp_on_dbs_scores[:, kd], paired=True)
    p_off_dopa_on_dbs = permutation_test(Xp_off_scores[:, kd], Xp_on_dbs_scores[:, kd], paired=True)
    p_on_dbs  = permutation_test(Xp_on_scores[:, kd], Xp_on_dbs_scores[:, kd], paired=True)

    p_double_off  = permutation_test(Xp_off_scores[:, kd], Xp_off_dbs_scores[:, kd], paired=True)
    p_on_dopa_off_dbs = permutation_test(Xp_off_dbs_scores[:, kd], Xp_on_scores[:, kd], paired=True)

    # --- FDR correction ---
    raw_pvals = [p_off_on, p_off_dbs, p_on_dbs, p_off_dopa_on_dbs] #, p_double_off, p_on_dopa_off_dbs
    reject, pvals_fdr, _, _ = multipletests(raw_pvals, alpha=0.05, method='fdr_bh')
    p_off_on_fdr, p_off_dbs_fdr, p_on_dbs_fdr, p_off_dopa_on_dbs_fdr = pvals_fdr #, p_double_off_fdr, p_on_dopa_off_dbs_fdr

    
    # --- Remove spines and grid ---
    sns.despine(ax=ax)
    ax.grid(False)  # completely remove the grid background
    
    # --- Annotate significance stars ---
    y_max = data.max()
    step = (data.max() - data.min()) * 0.08
    comparisons = [
        (0, 1, p_off_on_fdr),
        # (1, 2, p_on_dopa_off_dbs_fdr),
        (2, 3, p_off_dbs_fdr),
        # (0, 2, p_double_off_fdr),
        (1, 3, p_on_dbs_fdr),
        (0, 3, p_off_dopa_on_dbs_fdr)      
        
    ]
    print(comparisons)
    c = 1
    for i, (x1, x2, p) in enumerate(comparisons):
        if p < 0.05:
            y = y_max + (i+1*c) * step
            c = c+1
            ax.plot([x1, x1, x2, x2], [y, y, y, y], c='black', lw=1)
            ax.text((x1+x2)/2, y + step*0.5, get_sigstar(p),
                    ha='center', va='bottom', fontsize=18, fontweight='bold')

    # --- Title / label ---
    # ax.set_title(RPCs[kd], fontsize=15, pad=10)
    
    # --- Save publication-quality figure ---
    plt.tight_layout()
    os.makedirs('Figures/Main/Figure 2', exist_ok=True)
    plt.savefig(f"Figures/Main/Figure 2/Subpane 2C_scores{RPCs}.png", 
            dpi=1000, bbox_inches="tight", facecolor = 'white')   
    plt.show()
```
